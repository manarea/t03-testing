### Función Desarrollada Por un Alumno la cual quieremos revisar

```python
def fun_alumno(value: str) -> int:
    if value == 'true':
        return 0.   # <= BUG
    if value == 'false':
        return 1

    return int(value)

```

### Función Desarrollada por el Profesor: El Oraculo

```python
def fun_profesor(value: str) -> int:
    if value == 'true':
        return 1
    if value == 'false':
        return 0

    return int(value)
```

### Probando el DCCTester

```python
tester = DCCTester()
seed_input =  "false"
suite = tester.execute(seed_input, fun_alumno,fun_profesor)

counter = 1
for test in suite.gen_tests():
    print(f"test_{counter}: {test.test_result()} = \t input: {test.test_input()} \t ,output: {test.current_output()}\t ,expected: {test.test_expected_result()}")
    counter +=1
```

### Probando el HTML report

```python
from pygments import highlight
from pygments.lexers import PythonLexer
from pygments.formatters import HtmlFormatter
from IPython.core.display import HTML

def highlight_code(code):
    formatter = HtmlFormatter(style="colorful", full=True, cssclass="codehilite")
    highlighted_code = highlight(code, PythonLexer(), formatter)
    return HTML(highlighted_code)

result= TestResult(suite,fun_alumno)

highlight_code(result.full_report())
```
